```python
def isAngram(s,t):
    if len(s) != len(t):
        return False
    count = {}
    for ch in s:
        count[ch] = count.get(ch,0)+1
    for ch in t:
        count[ch] -= 1
    if count[ch] < 0:
        return False
    return True
print(isAngram("silent", "listen"))
```

```python
def arealmostEqual(s1,s2):
    if s1 == s2:
        return True
    diff =[]
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            diff.append(i)
        if len(diff) != 2:
            return True
    i,j = diff
    return s1[i] == s2[j] and s2[i] == s1[j]
print(arealmostEqual("banlk", "kanb"))
```

```python
#Check whether the number is palindrom or not with recursion
def palindrome(n, rev=0):
    if n == 0:
        return rev
    return palindrome(n // 10, rev * 10 + n % 10)

num = int(input("Enter a number: "))

if num == palindrome(num):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#codingbat sum67
def sum67(arr):
    total = 0
    skip = False

    for num in arr:
        if skip:
            if num == 7:
                skip = False
        else:
            if num == 6:
                skip = True
            else:
                total += num

    return total
```

```python
def sum67(arr, i=0):
    # Base case
    if i >= len(arr):
        return 0

    
    if arr[i] == 6:
        i += 1
        while arr[i] != 7:
            i += 1
        return sum67(arr, i + 1)

    
    return arr[i] + sum67(arr, i + 1)
```
